# Working with random data in dimensions.


```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
# n = 1000, 10000, 100000
# F.C = [60  -  200]
# F.R = [4   - 30]
# Temp = [35.0  -  36]
# T/A = [100 - 220 / 220 - 100]
# Glucosa = [30  - 450]
```

```python
FC = []
FR = []
Temp = []
TA = []
Gluc = []

```

```python
# For n = 1000
for _ in range(1000):
    FC.append(random.randint(60, 200))
    FR.append(random.randint(4, 30))
    Temp.append(random.uniform(35.0, 36.0))
    TA.append((random.randint(100,220)) / random.randint(20, 100))
    Gluc.append(random.randint(30, 450))
    
```

```python

len(TA)
```

```python
data = pd.DataFrame({'F.C': FC,
                    'F.R': FR,
                    'Temp': Temp,
                     'T/A': TA,
                     'Glucosa': Gluc
   })
```

```python
data
```

```python
# Crear una figura 3D
fig = plt.figure(figsize=(10, 8))  # Ajusta el tamaÃ±o de la figura

ax = fig.add_subplot(111, projection='3d')

# Crear el grÃ¡fico de dispersiÃ³n con colores diferentes para cada variable
scatter_FC = ax.scatter(data['F.C'], data['F.R'], data['Temp'], c='r', marker='o', label='F.C')
scatter_FR = ax.scatter(data['F.C'], data['F.R'], data['Temp'], c='g', marker='o', label='F.R')
scatter_Temp = ax.scatter(data['F.C'], data['F.R'], data['Temp'], c='b', marker='o', label='Temp')
scatter_Glucosa = ax.scatter(data['F.C'], data['F.R'], data['Temp'], c='y', marker='o', label='Glucosa')

# Agregar etiquetas de ejes
ax.set_xlabel('F.C')
ax.set_ylabel('F.R')
ax.set_zlabel('Temp')

# Agregar un tÃ­tulo
ax.set_title('GrÃ¡fico 3D de F.C, F.R y Temp con colores diferentes por variable')

# Mostrar la leyenda
ax.legend()

# Mostrar el grÃ¡fico
plt.show()




```
