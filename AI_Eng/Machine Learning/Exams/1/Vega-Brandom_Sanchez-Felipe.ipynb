## EXAMEN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
import sklearn.cluster as cluster
import sklearn.model_selection as training
import sklearn.metrics as metrics
from yellowbrick.cluster import KElbowVisualizer
from yellowbrick.cluster import silhouette_visualizer
```

```python
iris = datasets.load_iris()
iris = pd.DataFrame(iris.data, columns = iris.feature_names)
iris.head()
```

```python
X = iris.values
kkframe = iris[['sepal length (cm)', 'petal length (cm)', 'petal width (cm)']]
neded = kkframe.values
neded
```

## PCA con tres valores

```python
from sklearn.decomposition import PCA
# El profe de las 14-d quiere solo 2-d
pca = PCA(n_components = 1)

d1 = pca.fit_transform(neded)
d1
```

```python
newX = [iris[['sepal width (cm)']].values, d1]
newX = np.column_stack((iris[['sepal width (cm)']].values, d1))

newX
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 3, n_init = 'auto', random_state = 42)
y_kmeans = kmeans.fit_predict(newX)
y_kmeans
```

```python
colors = ['gray','brown', 'yellow']

plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 60, c = colors[2], label = 'Virginica')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 60, c = colors[1], label = 'Setosa')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 60, c = colors[0], label = 'Versicolor')

    
plt.title("Clulsters of Customers")
plt.ylabel("Sepal width")
plt.legend()
plt.show()
```

## Herarquico

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

dendograma = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendograma')

plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
# Plot clusters


plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 50, c = 'red', label = 'Virginica')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 50, c = 'blue', label = 'Setosa')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 50, c = 'green', label = 'Versicolor')
plt.title("Clusters of costomers")

plt.ylabel("Sepal width")
plt.legend()
plt.show()
```
