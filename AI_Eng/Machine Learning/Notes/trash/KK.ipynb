```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

```python
salaries_dataset = pd.read_csv('Salary.csv')
salaries_dataset.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = salaries_dataset.iloc[:, :-1].values
y = salaries_dataset.iloc[:, -1]. values
```

```python
plt.scatter(X, y)
plt.title("Salaries vs years experience")
plt.xlabel("years")
plt.ylabel('salary')


plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split( 
    X, y, test_size=0.2, random_state=101) 
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, linear_regression.predict(X_test), color = 'blue')
plt.title('Salary vs years of experience')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, linear_regression.predict(X_train), color = 'blue')
plt.title('Salary vs years of experience')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
