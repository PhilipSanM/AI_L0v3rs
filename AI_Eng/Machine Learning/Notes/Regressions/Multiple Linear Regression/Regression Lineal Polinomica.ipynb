# Regresion Lineal Polinomica

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("Position_Salaries.csv")
dataframe
```

```python
X = dataframe.iloc[:, 1:-1].values
y = dataframe.iloc[:, -1].values
```

```python
# The Data Set es muy pequeño pero se puede trabajar
# La variable dependiente es el salario...
# Metodos numericos y sino Ecuaciones Diferencial y ya sino modelo de IA
```

```python
# Esta muy pequeno, tonces como se hace :v, no va a haber split
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
# Training the polynomial Regression model on the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2) # degree of polynomio
X_poly = poly_reg.fit_transform(X)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Data:

```python
plt.scatter(X, y)
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

## Regresion Lineal simple

```python
plt.scatter(X, y)
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

## Regresion Lineal Polinomica

```python
plt.scatter(X, y)
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

El transform, eleva los datos de X al cuadrado y se los pasa.

Normalizar es pasarlos de 0 a 1, "Escalar".

Regresion polinomial trabaja mas facil con modelos de 0 a 1

```python
# ESCALADO:
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y)
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'red')
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
