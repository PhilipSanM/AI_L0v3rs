# Apriori

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Market_Basket_Optimisation.csv", header=None)
df.head()
```

```python
df.stack().value_counts(normalize=True)[:10].plot(kind='bar', title='Relative Frequency')
```

```python
df.stack().value_counts().apply(lambda item : item/df.shape[0])[:10].plot(kind='bar', title='Frequency')
```

```python
transactions =[]
for i in range (df.shape[0]):
  row = df.iloc[i].dropna().tolist()
  transactions.append(row)
```

```python
transactions[:10]
```

```python
!pip install apyori
```

```python
from apyori import apriori
rules = apriori(transactions= transactions, min_support= 0.004, min_length= 2, max_length=0)
results= list(rules)
results
```
